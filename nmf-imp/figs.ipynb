```python
import json
from pprint import pprint

with open('news24.json') as data_file:    
    data = json.load(data_file)
```

```python

def find_leaves(x):
    try: 
        for c in x['children']:
            find_leaves(c)
    except:
        print x['name'] + "\n\n\n"
```

```python
find_leaves(data)
```
